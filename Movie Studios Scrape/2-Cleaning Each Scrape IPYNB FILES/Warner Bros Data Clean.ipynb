# Warner Bros Data Cleaning from Scrapy Scrape

```python
import pandas as pd
```

```python
df = pd.read_csv('./warner_bros.csv')
```

```python
df
```

# Cleaning up the data

```python
df['year released']
```

```python
years_expanded = df['year released'].str.split(',', expand=True)
```

RESHAPE DATA FOR MERGE!

```python
a = years_expanded.T
a
```

```python
movie_name_expanded = df['movie name'].str.split(',',expand=True)
```

```python
b = movie_name_expanded.T
b
```

```python
world_wide = df['worldwide box'].str.split("$", expand=True)
```

```python
world_wide
c=world_wide.T
c.drop(0,inplace=True)
```

```python
c.reset_index(inplace=True)
c
```

```python
domestic = df['domestic box'].str.split("$", expand=True)
```

```python
d = domestic.T
```

```python

d.drop(0,inplace=True)
```

```python
d.reset_index(inplace=True)
```

```python
d
```

```python
international = df['international box'].str.split("$", expand=True)
```

```python
e=international.T
```

```python
e.drop(0,inplace=True)
```

```python
e.reset_index(inplace=True)
```

```python
e
```

# Time to merge!

Doesn't hurt to test it out!

```python
test = pd.concat([a,b], axis=1)
```

```python
test
```

```python
test2 = pd.concat([a,b,c,d,e], axis=1)
```

```python
test2
```

```python
pd. set_option('display.max_rows', None)
```

```python
warner_bros_csv = test2
```

Success! Let's Clean this up.

```python
warner_bros_csv
```

Remove unwanted index columns

```python
del warner_bros_csv['index']
```

```python
warner_bros_csv
```

RENAME OUR COLUMNS

```python
warner_bros_csv.columns=['year','movie name','worldwide box office','domestic box office','international box office']
```
