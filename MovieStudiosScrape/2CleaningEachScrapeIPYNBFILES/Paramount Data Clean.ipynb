# Cleaning Paramount Scrape from Scrapy

NOTE THIS WILL NOT RUN BECAUSE THE FOLDERS HAVE BEEN ORGANIZED. THIS IS JUST TO SEE THE STEPS OF THE WORK

```python
import pandas as pd
```

```python
df = pd.read_csv('./paramount_pictures.csv')
```

Scrapy returned a Dictionary of Lists So we Have to get in here and fix that up nicely!

```python
df
```

# Time to Clean!!!

```python
df['year released']
```

Seperate by comma values and expand

```python
years_expanded = df['year released'].str.split(',', expand=True)
```

This returns 100 columns instead of one...lets rotate!

```python
a = years_expanded.T
a
```

Rinse and Repeat!!

```python
movie_name_expanded = df['movie name'].str.split(',',expand=True)
```

```python
b = movie_name_expanded.T
b
```

```python
world_wide = df['worldwide box'].str.split("$", expand=True)
```

```python
world_wide
c=world_wide.T
c.drop(0,inplace=True)
```

```python
c.reset_index(inplace=True)
c
```

```python
domestic = df['domestic box'].str.split("$", expand=True)
```

```python
d = domestic.T
```

```python

d.drop(0,inplace=True)
```

```python
d.reset_index(inplace=True)
```

```python
d
```

```python
international = df['international box'].str.split("$", expand=True)
```

```python
e=international.T
```

```python
e.drop(0,inplace=True)
```

```python
e.reset_index(inplace=True)
```

```python
e
```

# MERGING!

```python
test = pd.concat([a,b], axis=1)
```

```python
test
```

```python
test2 = pd.concat([a,b,c,d,e], axis=1)
```

```python
test2
```
