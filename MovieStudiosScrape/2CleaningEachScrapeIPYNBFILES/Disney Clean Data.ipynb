Import pandas and let's get to work

# Time to Clean

Clean out the comma and use that to expand

```python
years_expanded = df['year released'].str.split(',', expand=True)
```

This will return 1 row so we'll Transpose it for shaping purposes

```python
a = years_expanded.T
```

```python
a.reset_index()
```

```python
movie_name_expanded = df['movie name'].str.split(',',expand=True)
```

We are going to have to drop values 40 and 53 because they had commas in the title and broke our shape of 100 columns.

```python
###drop 40 ###drop 53
```

```python
b = movie_name_expanded.T
b.drop(40,inplace=True)
```

```python
b.drop(53, inplace=True)
```

```python
b.reset_index(inplace=True)
```

```python
b
```

```python
world_wide = df['worldwide box'].str.split("$", expand=True)
```

```python
world_wide
c=world_wide.T
c.drop(0,inplace=True)
```

```python
c.reset_index(inplace=True)
len(c)
```

```python
domestic = df['domestic box'].str.split("$", expand=True)
```

```python
d = domestic.T
```

```python
len(d)
```

```python

d.reset_index(inplace=True)
```

```python
d.drop(0,inplace=True)
```

```python
d.reset_index(inplace=True)
```

```python
len(d)
```

```python
international = df['international box'].str.split("$", expand=True)
```

```python
e=international.T
```

```python
e.drop(0,inplace=True)
```

```python
len(e)
```

```python
e.reset_index(inplace=True)
```

```python
e
```
